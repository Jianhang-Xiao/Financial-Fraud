```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('PS_20174392719_1491204439457_log.csv')
```

```python
df_sample = df.sample(n=50000, random_state=42)
```

```python
# check how many missing values we have

df_sample.isna().sum()
```

```python
# create dictionary of mappings
rename_dict = {
    "step": "step", 
    "type": "type", 
    "amount": "amount",
    "nameOrig": "nameOrig",
    "oldbalanceOrg": "oldbalanceOrg",
    "newbalanceOrig": "newbalanceOrig",
    "isFraud": "isFraud",
    "isFlaggerFraud": "isFlaggerFraud"
}

# rename columns
df_sample_rename = df_sample.rename(columns=rename_dict)

# view new columns
df_sample_rename.columns
```

```python
selected = ["step", "nameOrig", "nameDest", "isFlaggedFraud"] #isFlaggedFraud is 0, so no need to include in the dataset.

df_sample_drop = df_sample_rename.drop(columns=selected)
```
